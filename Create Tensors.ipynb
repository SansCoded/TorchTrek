#Creating a Tensor -

```
# using empty - (2,3) is the size i.e 2/3 array

a = torch.empty(2,3)
```

```
# check type

type(a)
```

```
# using zeros -all values will be 0

torch.zeros(2,3)
```

```
# using ones -all values will be 1

torch.ones(2,3)
```

```
# using rand -created tensor with random values

torch.rand(2,3)
```

```
# use of seed - If u want the same random matrix each time u run use seed

torch.manual_seed(100)
torch.rand(2,3)
```

```
# using tensor - can customize tensor values

torch.tensor([[1,2,3],[4,5,6]])
```

```
# other ways

# arange - u can provide and step if u want to jump ex-(0,10)is range and 2 is step

print("using arange -", torch.arange(0,10,2))

#using linspace -linspace = linearly space..here u can give range like (0,10) and no. of values u want in the range like 10
# -so we got 10 linearly spaced values in range of 0,10

print(" ")
print("using linspace -",torch.linspace(0,10,10))

#using eye - here the diagonal items will be 1 and 5 is the size

print(" ")
print("using eye -", torch.eye(5))

#using full - (3,3) is size and 5 is the value

print(" ")
print("using full -", torch.full((3,3), 5))
```
