# Copying a Tensor -

```
a = torch.rand(2,3)
a
```

```
#using this u create exact copy-even after making changes to origninal it will reflect in copied tensor
b=a
b
```

```
#see as i make changes in a same will reflect in b
a[0][0]=0
print(a)
print(b)
```

```
# this is not feasible all the time as b is just another pointer pointing to a (same memory location)
print("A= ",id(a))
print("B= ",id(b))
```

```
#hence we can use function - clone

b = a.clone()
print(a)
print(b)
```

```
a[0][0] = 10
a
```

```
#this is because by using clone it has created a copy on another menmory location
print(b)
print("A= ",id(a))
print("B= ",id(b))
```
